## Line graph using pyplot function

```python
# Import all required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## Create dataframe

```python
lst1=list(range(1,10))
lst2=list(range(20,30))
df=pd.DataFrame(list(zip(lst1, lst2)),columns=['Number', 'val'])
```

## Plot dataframe using plot function 

```python
plt.plot(df.Number,df.val)
```

## How to add title and labels to graph?

```python
plt.plot(df.Number,df.val)
plt.title('Line chart example')
plt.xlabel('Number')
plt.ylabel('Value')
```

### modify text properties of graph's axis labels and title 

```python
import matplotlib.font_manager as font_manager
title_font = {'fontname':'Arial', 'size':'20', 'color':'black', 'weight':'normal','verticalalignment':'bottom'} # Bottom vertical alignment for more space
axis_font = {'fontname':'Arial', 'size':'16','color':'red'}
plt.plot(df.Number,df.val)
plt.xlabel("x axis", **axis_font)
plt.ylabel("y axis", **axis_font)
plt.title("Title", **title_font)
```

## How to change graph properties 
***
1) line color
***
The supported color abbreviations are the single letter codes

character	color
------------------------
1. 'b'  =	blue
2. 'g'	=   green
3. 'r'	=   red
4. 'c'	=   cyan
5. 'm'	=   magenta
6. 'y'	=   yellow
7. 'k'	=   black
8. 'w'	=   white

<p>full names ('green') or hex strings ('#008000').</p>


```python
plt.plot(df.Number,df.val, color='green')
```

***
2) marker for line point 
***
we can change marker for line plot using below input values

|character | description|
| --- | --- |
|'.'	|point marker
|','	|pixel marker|
|'o'	|circle marker|
|'v'	|triangle_down marker|
|'^'	|triangle_up marker|
|'<'	|triangle_left marker|
|'>'	|triangle_right marker|
|'1'	|tri_down marker|
|'2'	|tri_up marker|
|'3'	|tri_left marker|
|'4'	|tri_right marker|
|'s'	|square marker|
|'p'	|pentagon marker|
|'*'	|star marker|
|'h'	|hexagon1 marker|
|'H'	|hexagon2 marker|
|'+'	|plus marker|
|'x'	|x marker|
|'D'	|diamond marker|
|'d'	|thin_diamond marker|
|'|'	|vline marker|
|'_'	|hline marker|

```python
plt.plot(df.Number,df.val, marker='1', markersize=10)
```

***
3) Line Style and Line width
***
<p> We can change line style by using below input </p> 

character	description
--------------------------------
'-'	 = solid line style

'--' =	dashed line style

'-.' =	dash-dot line style

':'	 =dotted line style

```python
plt.plot(df.Number,df.val, color='green', marker='o', linestyle='-',linewidth=2, markersize=12)
```

## We can combine input formatting as one string to get same output

In below case

go-- => 
* g = green color
* o = circle
* -- = double dashed line

```python
plt.plot(df.Number,df.val,'go--')
```
