```python
# Import all required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

## Create Dataframe

```python
lst1=list(range(1,10))
lst2=list(range(20,30))
lst3=list(range(40,50))
df=pd.DataFrame(list(zip(lst1, lst2,lst3)),columns=['Number', 'val1','val2'])
```

## Create Subplot

```python
# Just a figure and one subplot
f, ax = plt.subplots()
ax.plot(df['Number'],df['val1'])
ax.set_title('Simple plot')
```

## Two subplots, the axes array is 1-d

```python
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(df['Number'],df['val1'])
axarr[0].set_title('Sharing X axis')
axarr[1].scatter(df['Number'],df['val2'])
```

## Two subplots, unpack the axes array immediately

```python
# Two subplots, unpack the axes array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(df['val1'],df['Number'])
ax1.set_title('Sharing Y axis figure 1')
ax2.scatter(df['val2'],df['Number'])
ax2.set_title('Sharing Y axis figure 2')
```

```python
    # row and column sharing
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(df['Number'],df['val1'])
ax1.set_title('Sharing x per column, y per row')
ax2.scatter(df['Number'],df['val2'])
ax3.scatter(df['Number'], 2 * df['val1'] ** 2 - 1, color='r')
ax4.plot(df['Number'], 2 * df['val1'] ** 2 - 1, color='r')
```
